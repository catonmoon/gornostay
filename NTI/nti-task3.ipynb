```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

```python
def norm(word_str):
    norm_ret = ''
    for word in word_str.split():
        p = morph.parse(word)[0]
        norm_ret += ' ' + p.normal_form
    return norm_ret
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from time import time
```

```python
clf = MultinomialNB()
```

```python
print("Training: ")
print(clf)
t0 = time()
train_text_clf = clf.fit(train_text, y_train.values)
train_title_clf = clf.fit(train_title, y_train.values)
train_time = time() - t0
print("train time: %0.3fs" % train_time)
```

```python
pred1 = np.argmax((train_title_clf.predict_proba(test_title) + train_text_clf.predict_proba(test_text)),1)
```

```python
pred2 = np.argmax((train_title_clf.predict_proba(test_title) * 0.5 + train_text_clf.predict_proba(test_text)),1)
```

```python
pred3 = np.argmax((train_title_clf.predict_proba(test_title) * 0.1 + train_text_clf.predict_proba(test_text)),1)
```
